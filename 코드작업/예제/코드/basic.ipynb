## 5. Autograd

Back Propagation은 딥러닝 모델이 학습하는 핵심요소이다. 이 방법은 pytorch에서 Autograd 방식으로 구현 할 수 있다. 다음 예제를 보며 pytorch에서 Back Propagation 구현을 어떻게 하는지 알아보자.

### device

행렬 연산을 수행할 때, 병렬연산을 수행하기 위해 GPU를 사용할 수 있다. GPU를 사용하기 위해서는 cuda와 cuDNN이 설치가 되어 있어야 하며, 그래픽 카드에 맞는 버전의 cuda version이 설치가 되어 있어야 GPU를 연동 할 수 있다. GPU를 사용 가능한지는 `torch.cuda.is_available()`함수로 확인 할 수 있다.

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") # gpu가 있으면 gpu를, 없으면 cpu 쓰겠다는 코드
device
```

BATCH_SIZE : 파라미터가 업데이트 될 때 계산되는 데이터의 개수<br/>
INPUT_SIZE : 딥러닝 모델에서 입력데이터의 컬럼 숫자.<br/>
HIDDEN_SIZE : Input을 다수의 파라미터를 이용해 계산한 결과에 한번더 계산되는 파랍미터의 수. ex) \[2, 3(INPUT_SIZE)\] @ \[3, 2(HIDDEN_SIZE)\] -> \[2, 2\] <br/>
OUTPUT_SIZE : 딥러닝 모델에서 최종적으로 출력되는 값의 벡터의 크기 HIDDEN_SIZE와 같은 개념.

```python
BATCH_SIZE = 64
INPUT_SIZE = 1000
HIDDEN_SIZE = 100
OUTPUT_SIZE = 10
```

```python
x = torch.randn(BATCH_SIZE,
               INPUT_SIZE,
               dtype=torch.float32, 
               device=device,
               requires_grad=False) # [64. 1000]
y = torch.randn(BATCH_SIZE,
               OUTPUT_SIZE,
                dtype=torch.float32,
                device=device, 
               requires_grad=False) # [64, 10]
w1 = torch.randn(INPUT_SIZE,
                HIDDEN_SIZE,
                dtype=torch.float32,
                device=device,
                requires_grad=True) # [1000, 100]
w2 = torch.randn(HIDDEN_SIZE,
                OUTPUT_SIZE,
                dtype=torch.float32,
                device=device,
                requires_grad=True) # [100, 10]
```

### Gradient 계산
$$ w_l = w_{l-1} - \eta\Delta J(w_{l-1})$$

$w_l$ : gradient가 계산된 후의 학습된 파라미터

$\eta$ : learning rate

$J(\odot)$ : $w_{l-1}$ 에대한 미분 - Jacobian matrix

```python
learning_rate = 1e-6
for t in range(1, 501): # 500 epoch
    y_pred = x.mm(w1).clamp(min=0).mm(w2) # Relu activiation function을 사용해 비선형성을 준 MLP 모델
    
    loss = (y_pred - y).pow(2).sum() # loss function - squred error R^2 loss
    if t % 100 == 0:
        print("iteration: ", t, "\t", "loss : ", loss.item()) # print loss
    loss.backward() # backpropagation
    
    with torch.no_grad():
        w1 -= learning_rate * w1.grad # w1 parameters 업데이트
        w2 -= learning_rate * w2.grad # w2 parameters 업데이트
        
        w1.grad.zero_() # w1 gradient 초기화
        w2.grad.zero_() # w2 gradient 초기화
```

# Multi Layer Perceptrons(MLP) with LogisticsBinaryClassifier

LogisticsBinaryClassifier는 Binary Classification에 사용할 수 있으다. 분류는 0,1로 sigmoid function을 이용해 0.5 이상이면 1, 그 이하는 0으로 집합을 구분할 수 있다. sigmoid function을 1일 확률로 볼 수 있다.

## log-odd
**odd ratio** : 어떤 이벤트가 발생했을 때, 특정 특징을 보이는 그룹은 어떤 그룹인가? ex) 행사에서 술을 마시는 그룹은 어떤 그룹인가?


|              | **Drink** | **Don't Drink** | **Total** |
|:------------:|:---------:|:---------------:|-----------|
| **Students** |     90    |        10       |    100    |
| **Teachers** |     20    |        80       |    100    |
| **total**    |    110    |        90       |    200    |

학생이 마실 확률 : 90 / 100 = 0.9
학생이 마시지 않을 확률 : 0.1

성생님이 마시지 않을 확률 : 0.8

odd ratio = 9 / 0.4 = 36

# Loss Function

## Binary Cross Entropy

$$ BCE = -\frac 1 N \sum ylog(\hat{y}) + (1-y)log(1-\hat{y})$$

$$ BCE = -\frac 1 N \sum y log(\frac {\hat{y}}{(1-\hat{y})}) + log(1 -\hat{y})$$

```python
# Binary Cross Entropy
ln = np.linspace(1e-03, 1, 1000)

def BCE(x, y=1):
    return -(y * np.log(x) + (1-y)*np.log(1 - x))

f, ax = plt.subplots(1, 2, figsize=(10, 5))
# if y = 1
y1 = BCE(ln)
ax[0].plot(ln, y1)
# if y = 0
y2 = BCE(ln, y=0)
ax[1].plot(ln, y2)
plt.show()
```

```python
x1 = np.array([
    [0.1, 0.72],
    [0.35, 0.54],
    [0.45, 0.13],
    [0.6, 0.3]
])

x2 = np.array([
    [0.3, 0.96],
    [0.15, 1.],
    [0.7, 0.85],
    [0.95, 0.4]
])

y = np.array([1,1,1,1,0,0,0,0])

ablinex = np.linspace(0, 1, 1000)
abliney = np.linspace(1, 0, 1000)
plt.scatter(x1[:, 0],x1[:, 1], c="red", label="x1")
plt.scatter(x2[:, 0],x2[:, 1], c="blue", label="x2")
plt.plot(ablinex, abliney, c="black")
plt.legend()
plt.show()
```

참고 : sigmoid 함수 $sigmoid = \frac 1 {1 + e^{-x}}$

```python
# x1, 과 x2를 LogisticsClassifier 을 pytorch를 이용해 학습시켜보자.
dataset = np.concatenate([x1, x2], axis=0)
```

```python
# weight to train
w = torch.tensor(np.random.randn(2,1), requires_grad=True, device=device)
bias = torch.tensor(np.random.rand(1), requires_grad=True, device=device)
```

```python
def sigmoid(x):
    return 1/(1 + torch.exp(-x))

def criterion(y_pred, y):
    return -(y * torch.log(y_pred) + (1-y) * torch.log(1-y_pred))
```

```python
learning_rate = 1e-3

optimizer = torch.optim.Adam([w, bias], lr=learning_rate)

for epoch in range(1, 20000):
    losses = []
    for X, y_true in zip(dataset, y.reshape(len(y), 1)):
        X = torch.tensor(X, device=device)
        y_true = torch.tensor(y_true, device=device)
        y_pred =  sigmoid(torch.matmul(X, w) + bias)

        loss = criterion(y_pred, y_true) # loss function - binary cross entropy
        losses.append(loss)
        
        loss.backward() # backpropagation
        
        optimizer.step()
        optimizer.zero_grad()

    if epoch % 100 == 0:
        print("iteration: ",epoch, "\t", "loss : ", torch.mean(torch.tensor(losses)).item()) # print loss
    if loss.item() < 0:
        break
```

```python
with torch.no_grad():
    predicts = []
    for X in dataset:
        X = torch.tensor(X, requires_grad=False, device=device)
        predict = sigmoid(torch.matmul(X, w) + bias)
        predict = predict.to("cpu").numpy()
        predicts.append(predict)

predicts = np.asarray(predicts)
    
predicts = np.where(predicts >= 0.5, 1, 0).reshape(8)

x = np.concatenate([x1, x2], axis=0)

f, axes = plt.subplots(1, 2)

axes[0].scatter(x1[:, 0], x1[:, 1], c="red", label="x1")
axes[0].scatter(x2[:, 0], x2[:, 1], c="blue", label="x2")
axes[0].set_title("answer") 
axes[0].legend()

axes[1].scatter(x1[:, 0], x1[:, 1], c=["red" if p == 1 else "blue" for p in predicts[:4]], label="x1")
axes[1].scatter(x2[:, 0], x2[:, 1], c=["red" if p == 1 else "blue" for p in predicts[4:8]], label="x2")
axes[1].set_title("predict")
axes[1].legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import transforms, datasets
import torchmetrics.functional as MF
```

```python
# device 확인
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
BATCH_SIZE = 32
EPOCHS = 10
```

```python
# load dataset
train_dataset = datasets.FashionMNIST(root="data/FashionMNIST", train=True, download=True, transform = transforms.ToTensor())
test_dataset = datasets.FashionMNIST(root="data/FashionMNIST", train=False, download=True, transform = transforms.ToTensor())
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python
for (X_train, y_train) in train_loader:
    print(f"X_train: {X_train.size()} type: {X_train.type()}")
    print(f"y_train: {y_train.size()} type: {y_train.type()}")
    break
```

```python
# 어떤 데이터들이 있는지 확인 해 보기
pltsize= 2
f, axes = plt.subplots(1, 10, figsize=(10 * pltsize, pltsize))
for i in range(10):
    axes[i].axis('off')
    axes[i].imshow(X_train[i, :, :, :].numpy().reshape(28, 28), cmap="gray_r")
    axes[i].set_title(f"Class: {y_train[i].item()}")
```

```python
# Auto encoder
class AE(nn.Module):
    def __init__(self):
        super(AE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(28 * 28, 512),
            nn.ReLU(),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, 32)
        )
        self.decoder = nn.Sequential(
            nn.Linear(32, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, 28 * 28)
        )
        
    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return encoded, decoded
```

```python
model = AE().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss()

print(model)
```

```python
def train(model, train_loader, optimizer, epoch, log_interval):
    model.train()
    for batch_idx, (image, _) in enumerate(train_loader):
        image = image.view(-1, 28 * 28).to(device)
        target = image.view(-1, 28 * 28).to(device)
        optimizer.zero_grad()
        encoded, decoded = model(image)
        loss = criterion(decoded, image)
        loss.backward()
        optimizer.step()
        
        if batch_idx % log_interval == 0:
            print("Train Epoch: {} [{}/{}({:.0f}%)]\tTrain loss: {:.6f}".format(
                    epoch, 
                    batch_idx * len(image), 
                    len(train_loader.dataset), 
                    100 * batch_idx / len(train_loader), 
                    loss.item()
                )
            )
def evaluate(model, test_loader):
    model.eval()
    test_loss = 0
    real_image = []
    gen_image = []
    with torch.no_grad():
        for image, _ in test_loader:
            image = image.view(-1, 28 * 28).to(device)
            target = image.view(-1, 28 * 28).to(device)
            encoded, decoded = model(image)
            test_loss += criterion(decoded, image).item()
            real_image.append(image.to('cpu'))
            gen_image.append(decoded.to('cpu'))
    test_loss /= len(test_loader.dataset)
    return test_loss, real_image, gen_image
```

```python
%matplotlib inline

for epoch in range(1, EPOCHS + 1):
    train(model, train_loader, optimizer, epoch, log_interval = 200)
    test_loss, real_image, gen_image = evaluate(model, test_loader)
    print("\n[EPOCH]:{}], \tTest Loss: {:.4f}".format(epoch, test_loss))
    f, a = plt.subplots(2, 10, figsize=(10, 4))
    for i in range(10):
        img = np.reshape(real_image[0][i], (28, 28))
        a[0][i].imshow(img, cmap="gray_r")
        a[0][i].set_xticks(())
        a[0][i].set_yticks(())
        
    for i in range(10):
        img = np.reshape(gen_image[0][i], (28, 28))
        a[1][i].imshow(img, cmap="gray_r")
        a[1][i].set_xticks(())
        a[1][i].set_yticks(())
    plt.show()
```

```python
def get_embeddings(model, test_loader):
    model.eval()
    embeddings = []
    Ys = []
    with torch.no_grad():
        for image, y in test_loader:
            image = image.view(-1, 28 * 28).to(device)
            encoded, decoded = model(image)
            embeddings.append(encoded.to('cpu').numpy())
            Ys.append(y.numpy())
            
    embeddings = np.concatenate(embeddings, axis=0)
    Ys = np.concatenate(Ys)
    return embeddings, Ys
```
